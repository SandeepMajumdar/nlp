```python
import transformers
from transformers import pipeline
```

```python
classifier_db_base = pipeline("sentiment-analysis")
classifier_db_base("I'm sure I will be able to crack that interview")
```

```python
text = 'i want to study'
classifier_db_base(text)
```

```python
text = ['i love to eat', 'he is about to miss the bus', 'are we on the right track?']
classifier_db_base(text)
```

```python
classifier_fb_bart = pipeline("zero-shot-classification")
text = 'i want to study'
classifier_fb_bart(text, candidate_labels=['education', 'business', 'politics'])
```

```python
text = 'government is encouraging tax planning by corporates'
classifier_fb_bart(text, candidate_labels=['education', 'business', 'politics'])
```

```python
text = 'congress party is all set to return'
classifier_fb_bart(text, candidate_labels=['education', 'business', 'politics'])
```

```python
generator_gpt2 = pipeline('text-generation')
```

```python
text = 'I wish I got the'
generator_gpt2(text)
```

```python
generator_distilgpt2 = pipeline('text-generation', model='distilgpt2')
```

```python
text = 'I want to make it very clear that'
generator_distilgpt2(text, max_length=30, num_return_sequences=2)
```

```python
unmasker_distilroberta_base = pipeline('fill-mask')
```

```python
text = 'Do you know how much I <mask> you?'
unmasker_distilroberta_base(text, top_k=2)
```

```python
ner_dbmdz_bert = pipeline('ner', grouped_entities=True)
```

```python
text = 'My name is Sandeep and I work at DotAI in Mumbai'
ner_dbmdz_bert(text)
```

```python
for i in ner_dbmdz_bert(text):
  print('{}: \t {}'.format(i['entity_group'], i['word']))
```

```python
qna_db_base = pipeline('question-answering')
```

```python
question = 'Where do I work?'
context = 'My name is Sandeep and I work at DotAI in Mumbai'
qna_db_base(question=question, context=context)['answer']
```

```python
summarizer_ssh_dbart = pipeline('summarization')
```

```python
text = '''
Television is one of the many wonders of modern science and technology. It was invented in England by the Scottish scientist J.N. Baird 
in 1928 and the British Broadcasting Corporation was the first to broadcast television images in 1929. Previously the radio helped us 
hear things from far and near. spread information and knowledge from one corner of the globe to another. But all this was done through 
sound only. But television combined visual images with sound. Today we can watch games, shows, and song and dance programs from all 
corners of the world while sitting at our own homes. TV can be used for educating the masses, for bringing to us the latest pieces of 
information audio-visually and can provide us all kinds of entertainment even in color. But as in all things, too much televiewing may 
prove harmful. In many cases, the habit of watching TV has an adverse effect on the study habits of the young. When we read books, we 
have to use our intelligence and imagination. But in most cases, TV watching is a passive thing. It may dull our imagination and 
intelligence.
'''
summarizer_ssh_dbart(text)
```
